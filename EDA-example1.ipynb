```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('Laptop.csv')
```

```python
df.head()
```

**Lets do some visualization**

```python
df.groupby(['Brand']).count()['Model'].plot(kind = 'bar', grid = True, color ='mediumaquamarine', figsize = (12,8))
plt.show()
```

```python
df.groupby(['Processor']).count()['Model'].plot(kind = 'bar', grid = True, color ='mediumaquamarine', figsize = (12,8))
plt.show()
```

```python
df.info()
```

```python
df.groupby('Brand').mean()['Rating'].sort_values(ascending = False).plot(kind = 'barh', 
            color = 'mediumaquamarine', figsize = (12,8), 
            grid = True, fontsize = 15)
plt.ylabel('Brand', fontsize = 15)
plt.xlabel('Rating', fontsize = 15)
plt.title('Average Rating for each Brand', fontsize = 15)
plt.show()
```

```python
df.RAM.value_counts().plot(kind = 'bar', 
        color = 'mediumaquamarine', grid = True, figsize = (12,8), 
        fontsize = 15, rot =0)
plt.xlabel('OS', fontsize = 15)
plt.ylabel('Amount of OS', fontsize = 15)
plt.show()
```

```python
df[['Brand','Model','Rating']].groupby('Brand').mean().sort_values(by='Rating', ascending = False).head(10).plot(kind = 'barh', 
        color = 'mediumaquamarine', grid = True, figsize = (12,8), 
        fontsize = 15, rot =0, legend = None)
plt.xlabel('Average Rating', fontsize = 15)
plt.ylabel('Brand', fontsize = 15)
plt.title("Top 10 Average Brand's Model Rating", fontsize = 15)
plt.show()
```

```python
df['ssd_hdd'] = df['Hard_Disk_Capacity'].str[-3:]
df.groupby('ssd_hdd').count()['Model'].iloc[:2].plot(kind = 'barh', 
        color = 'mediumaquamarine', grid = True, figsize = (12,8), 
        fontsize = 15, rot =0, legend = None)
plt.xlabel('Amount of Hard Disks', fontsize = 15)
plt.ylabel('Hard Disk type', fontsize = 15)
plt.title("Amount of each Hard Disk type", fontsize = 15)
plt.show()
```
